```python
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(19, 8))
 
# set height of bar
Facecream = [2500,2630,2140,3400,3600,2760,2980,3700,3540,1990,2340,2900]
Facewash = [1500,1200,1340,1130,1740,1555,1120,1400,1780,1890,2100,1760]
Toothpaste = [5200,5100,4550,5870,4560,4890,4780,5860,6100,8300,7300,7400]
Bathing = [9200,6100,9550,8870,7760,7490,8980,9960,8100,10300,13300,14400]
Shampoo = [1200,2100,3550,1870,1560,1890,1780,2860,2100,2300,2400,1800]
Moist = [1500,1200,1340,1130,1740,1555,1120,1400,1780,1890,2100,1760]
 
# Set position of bar on X axis
br1 = np.arange(len(Facecream))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]
br5 = [x + barWidth for x in br4]
br6 = [x + barWidth for x in br5]

# Make the plot
plt.bar(br1, Facecream, color ='blue', width = barWidth,
        edgecolor ='grey', label ='Facecream')
plt.bar(br2, Facewash, color ='black', width = barWidth,
        edgecolor ='grey', label ='Facewash')
plt.bar(br3, Toothpaste, color ='green', width = barWidth,
        edgecolor ='grey', label ='Toothpaste')
plt.bar(br4, Bathing, color ='cyan', width = barWidth,
        edgecolor ='grey', label ='Bathing')
plt.bar(br5, Shampoo, color ='red', width = barWidth,
        edgecolor ='grey', label ='Shampoo')
plt.bar(br6, Moist, color ='yellow', width = barWidth,
        edgecolor ='grey', label ='Moist')
 
# Adding Xticks
plt.xlabel('Mois', fontweight ='bold', fontsize = 15)
plt.ylabel('Nombres unités vendues', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(Facecream))], ['Janvier', 'Fevrier', 'Mars', 'Avril', 'Mai', 'Juin', 'Juillet', 'Aout', 'Sept', 'Oct', 'Nov', 'Dec'])
 
plt.legend()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-12, 0, 0.25)
Y = np.arange(-6, 0, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
```
